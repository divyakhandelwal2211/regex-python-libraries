```python
import seaborn as sns
import matplotlib.pyplot as plt

height = [62, 64, 69, 75, 66,
68, 65, 71, 76, 73]

weight = [120, 136, 148, 175, 137,
165, 154, 172, 200, 187]

#plt.scatter
sns.scatterplot(x=height, y=weight)

plt.show()
```

```python
#bar and countplot (univariate)
gender = ["Female", "Female",
"Female", "Female",
"Male", "Male", "Male",
"Male", "Male", "Male"]

#Univariate analysis: categorical v.
sns.countplot(x=gender)

plt.show()
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df.smoker.unique()
```

```python
df.day.value_counts()
```

```python
#basic scatter plot
sns.scatterplot(x="total_bill",y="tip",data=df)

plt.show()
```

```python
#basic scatter plot
sns.scatterplot(x="total_bill",y="tip",data=df, hue="day")

plt.show()
```

```python
#basic scatter plot
sns.scatterplot(x="total_bill",y="tip",data=df, hue="smoker")

plt.show()
```

```python
#basic scatter plot
sns.scatterplot(x="total_bill",y="tip",data=df, hue="time")

plt.show()
```

```python
#basic scatter plot
sns.scatterplot(x="total_bill",y="tip",data=df, hue="time", hue_order=["Dinner", "Lunch"])

plt.show()
```

```python
#basic scatter plot
sns.scatterplot(x="total_bill",y="tip",data=df, hue="sex")

plt.show()
```

```python
hue_colors = {'Yes':'black',
              'No':'red'}


sns.scatterplot(x='total_bill',y='tip',data=df,hue='smoker',hue_order=['Yes','No'],palette=hue_colors)
plt.show()
```

```python
#countplot on categorical variables (univariate)
sns.countplot(x="smoker",
data=df, hue="sex")

plt.show()
```

```python
#countplot on categorical variables (univariate)
sns.countplot(x="time",
data=df, hue="day")

plt.show()
```

```python
#countplot on categorical variables (univariate)
sns.countplot(x="sex",
data=df,hue="day")

plt.show()
```

### Relational Plot Interface

```python
sns.relplot(x='total_bill',y='tip',data=df,kind='scatter')

plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,kind='scatter',hue='smoker')

plt.show()
```

```python
sns.relplot(x="total_bill",
y="tip",
data=df,
kind="scatter", col="smoker")

plt.show()
```

```python
sns.relplot(x="total_bill",
y="tip",
data=df,
kind="scatter", row="smoker")

plt.show()
```

```python
sns.relplot(x="total_bill",
y="tip",
data=df,
kind="scatter",
col="day")

plt.show()
```

```python
sns.relplot(x="total_bill",
y="tip",
data=df,
kind="scatter",
row="day")

plt.show()
```

```python
sns.relplot(x="total_bill",
y="tip",
data=df,
kind="scatter",
col="time",
row="day")

plt.show()
```

```python
sns.relplot(x='total_bill',y='tip'
            ,data=df,
            kind='scatter',
            col='day',
            col_wrap=3,
            col_order=['Fri','Thur','Sat','Sun'])

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.relplot(x="total_bill",
y="tip",
data=df,
kind="scatter",hue="time", size="day")
plt.show()
```

```python
# Changing point transparency (Set alpha to be between 0 and 1)


sns.relplot(x="total_bill",
y="tip",
data=df,
kind="scatter",
hue="day", alpha=0.5)

plt.show()
```

```python
#basic line plot
sns.relplot(x="total_bill",
y="tip",
data=df,kind="line")

plt.show()
```

```python
sns.relplot(x="total_bill",
y="tip",
data=df,kind="line",style="sex",hue="sex",markers=True)#style="day",hue="day",markers=True,dashes=False
#,hue="day",markers=True,dashes=False

plt.show()
```

```python
sns.relplot(x="day",
y="total_bill",
data=df,kind="line",errorbar=None)

plt.show()
```

```python
sns.countplot(x="smoker",
data=df)

plt.show()
```
