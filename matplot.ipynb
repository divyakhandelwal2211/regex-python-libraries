### Using oop styling

```python
import numpy as np
z = np.linspace(0,2,100)

fig,ax = plt.subplots()

ax.plot(z,z,label = 'linear')
ax.plot(z,z**2,label = 'quadratic')
ax.plot(z,z**3,label = 'cubic')

ax.set_xlabel('x label')
ax.set_ylabel('y label')

ax.set_title('simple plot')

ax.legend()

plt.show()

```

```python
import matplotlib.pyplot as plt
# x-axis values
a = [2,4,7,9,13]

# Y-axis values
b = [3,7,9,12,18]

# Z values
c = [6,7,5,8,15]

# Function to plot
plt.plot(a,b, marker="*", color="r", label="first plot")#1st line plot

plt.plot(b,c, marker="p", color="b", linestyle="-.", label="second plot") #2nd line plot

plt.plot(a,c, marker="+", color="g", linestyle = "--", label="third plot") #3rd line plot

plt.xlabel('Height')
plt.ylabel('age')

plt.title("Simple Plot")
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(3, 1,sharex=True)
```

```python
# x-axis values
a = [2,4,7,9,13]

# Y-axis values
b = [3,7,9,12,18]

# Z values
c = [6,7,5,8,15]

#assign a background style
plt.style.use('bmh')

fig, ax = plt.subplots(3, 1,sharex=True)

# applicable for single row or single column (important)
ax[0].plot(a, b, marker="v", color="r", linestyle="--")#first row

ax[1].plot(b, c, marker="o", color="b", linestyle="-.")#second row

ax[2].plot(a, c, marker="+", color="g", linestyle="-")#third row

ax[0].set_ylabel("nikhil")
ax[1].set_ylabel("garima")
ax[2].set_ylabel("harika")
ax[2].set_xlabel("Simple Plot")
ax[0].set_title("Grid Plots")

# function to show the plot
plt.show()
```

```python
#assign a background style
plt.style.use('seaborn-v0_8-dark')

#applicable only if you are working with a single column or a single row
fig, (ax1, ax2, ax3) = plt.subplots(3, 1,sharex=True)
#ax1 is first row
#ax2 is second row/grid
#ax3 is third row/grid

ax1.plot(a, b, marker="v", color="r", linestyle="--")

ax2.plot(b, c, marker="o", color="b", linestyle="-.")

ax3.plot(a, c, marker="+", color="g", linestyle="-")

ax1.set_ylabel("nikhil")
ax2.set_ylabel("garima")
ax3.set_ylabel("harika")
ax3.set_xlabel("Simple Plot")

# function to show the plot
plt.show()
```
