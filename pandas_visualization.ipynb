```python
import pandas as pd
import matplotlib.pyplot as plt

#pd.options.plotting.backend = 'plotly'
```

```python
data = {'Sales': [1000, 1500, 1200, 1800, 2000],
    'Expenses': [600, 700, 800, 1000, 1200]}

Year = ["2018", "2019", "2020", "2021", "2022"]

df = pd.DataFrame(data,index=Year)
df
```

```python
df.shape
```

```python
#side by side
df.plot(y=["Sales","Expenses"],kind="bar")
```

```python
df.plot(y=["Sales","Expenses"],kind="line",marker="+") #markers in plotly
```

```python
df.plot(kind="hist",bins=5,stacked=False,orientation="vertical")
```

```python
#histogram
df.plot(kind="hist",bins=5)
```

```python
color = {
    "boxes": "green",
    "whiskers": "orange",
    "medians": "blue",
    "caps": "red",
}
```

```python
#box plot
df.boxplot(column=["Sales","Expenses"],vert=True,color=color,positions=[1,5]) #true by default
```

```python
df["Sales"].plot(kind ="box",color=color)
```

```python
df["Sales"].plot(kind="pie", colors = ["blue","green","red","cyan","yellow"],autopct="%.0f")
```

```python
df.plot.pie(y="Sales",autopct="%.0f")
```

```python
df.plot(x='Sales', y='Expenses', kind='scatter', color='blue')
```

```python
df["Sales"].plot(kind="kde",color="red") #kernel density estimation  
```

```python
df.plot(kind="kde",stacked=True) #kernel density estimation
```

```python
df.plot(x="Sales",y="Expenses",kind="hexbin",gridsize=10,cmap="Oranges")
```
